```python
x=final.drop(['prognosis'],axis=1).values
```

```python
y=final['prognosis'].values
```

```python
from sklearn.model_selection import train_test_split           ## data separation
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import RobustScaler    #to handle outliers
```

```python
from sklearn.ensemble import RandomForestClassifier      ## importing randomforest
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
rfc=RandomForestClassifier(n_estimators=60,random_state=23)
```

```python
rfc.fit(x_train,y_train)    ## model training
```

```python
rfc.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score    ## used to calculate accuarcy of classification model
```

```python
y_pred=rfc.predict(x_test)      
```

```python
accuracy_score(y_pred,y_test)
```

```python
import sklearn.metrics   ## used to quantitative measures to evaluate models performance
```

```python
print(sklearn.metrics.classification_report(y_test,y_pred))
```
